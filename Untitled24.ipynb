```python
import pandas as pd

df = pd.read_csv("Student Performance.csv")
```

```python
random_sample = df.sample(n=150, random_state=42)
```

1. Descriptive Statistics

```python
import pandas as pd
import numpy as np

def descriptive_statistics(sample, field):
    desc = {}
    desc['Count'] = sample[field].count()
    desc['Mean'] = sample[field].mean()
    desc['Median'] = sample[field].median()
    desc['Standard Deviation'] = sample[field].std()
    desc['Minimum'] = sample[field].min()
    desc['Maximum'] = sample[field].max()
    desc['Variance'] = sample[field].var()
    desc['Skewness'] = sample[field].skew()
    desc['Kurtosis'] = sample[field].kurt()
    return pd.Series(desc)

random_desc = descriptive_statistics(random_sample, 'exam_score')
print("Descriptive Statistics for Random Sample (exam_score)")
print(random_desc)
```

2. Random Sampling

```python
random_sample = df.sample(n=150, random_state=42)

random_desc = descriptive_statistics(random_sample, 'exam_score')

print("Mean:", random_desc['Mean'])
print("Median:", random_desc['Median'])
print("Standard Deviation:", random_desc['Standard Deviation'])
```

3. Systematic Sampling

```python
k = len(df) // 150
systematic_sample = df.iloc[::k].head(150)

systematic_desc = descriptive_statistics(systematic_sample, 'exam_score')

print("number of total", k)
```

4. Visualizations (Scatter, Box, Histogram, Heatmap)

```python
import matplotlib.pyplot as plt

# Scatter Plot
plt.figure()
plt.scatter(df['study_hours'], df['exam_score'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')
plt.show()

# Box Plot
plt.figure()
plt.boxplot(df['exam_score'])
plt.title('Box Plot of Exam Score')
plt.ylabel('Exam Score')
plt.show()

# Histogram
plt.figure()
plt.hist(df['exam_score'], bins=20)
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.title('Histogram of Exam Score')
plt.show()

# Heatmap
plt.figure()
corr_matrix = df.corr(numeric_only=True)
plt.imshow(corr_matrix)
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.colorbar()
plt.title("Correlation Heatmap")
plt.show()
```

5. Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

df = pd.read_csv("Student Performance.csv")
df = df.dropna(subset=["study_hours", "exam_score"])

X = df[["study_hours"]]
y = df["exam_score"]

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
print(f"Equation: exam_score = {model.intercept_:.2f} + {model.coef_[0]:.2f} * study_hours")
print("R2 Score:", round(r2_score(y, y_pred), 3))

df_sorted = df.sort_values("study_hours")
X_sorted = df_sorted[["study_hours"]]
y_pred_sorted = model.predict(X_sorted)

plt.figure(figsize=(8,5))
plt.scatter(df["study_hours"], df["exam_score"], alpha=0.5)
plt.plot(df_sorted["study_hours"], y_pred_sorted)
plt.title("Scatter Plot with Linear Regression Line")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
6. Correlation (Pearson & Spearman + Chi-square)
```

```python
from scipy import stats

num_df = random_sample[["study_hours", "exam_score"]].dropna()

pearson_corr, pearson_p = stats.pearsonr(num_df["study_hours"], num_df["exam_score"])
spearman_corr, spearman_p = stats.spearmanr(num_df["study_hours"], num_df["exam_score"])

print("===== Numerical Correlation: Study Hours vs Exam Score =====")
print(f"Pearson Correlation: {pearson_corr:.3f}")
print(f"Pearson P-value: {pearson_p:.5f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
print(f"Spearman P-value: {spearman_p:.5f}")

# Categorical correlation
random_sample["Score_Category"] = pd.qcut(random_sample["exam_score"], q=3, labels=["Low", "Medium", "High"])

contingency_table = pd.crosstab(random_sample["gender"], random_sample["Score_Category"])

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\n===== Categorical Correlation: Gender vs Exam Score =====")
print(f"Chi-square Statistic: {chi2:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.5f}")

if p < 0.05:
    print("Significant relationship between Gender and Exam Score.")
else:
    print("No significant relationship between Gender and Exam Score.")
```

7. One Sample T-test

```python
from scipy.stats import ttest_1samp

t_stat, t_p = ttest_1samp(random_sample['exam_score'], 20)

print("T-statistic:", t_stat)
print("P-value:", t_p)
```
